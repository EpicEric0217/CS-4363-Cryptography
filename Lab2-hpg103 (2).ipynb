```python
import sys
import numpy as np


def cipher_encryption():
    mssg = input("Enter the message: ").upper()
    mssg = mssg.replace(" ", "")
    
    # Handle condition if the message length is odd.
    len_chkr = 0
    if len(mssg) % 2 != 0:
        mssg += "0"
        len_chkr = 1

    # Convert msg to matrices
    colm = int(len(mssg)/2)
    row = 2
    mssg2d = np.zeros((row, colm), dtype=int)
    
    iter1 = 0
    iter2 = 0
    for i in range(len(mssg)):
        if i % 2 == 0:
            mssg2d[0][iter1] = int(ord(mssg[i])-65)
            iter1 += 1
        else:
            mssg2d[1][iter2] = int(ord(mssg[i])-65)
            iter2 += 1
        
    key = input("Enter the 4 letter key string please: ").upper()
    key = key.replace(" ", "")

    # Convert key to 2x2
    key2d = np.zeros((2, 2), dtype=int)
    iter3 = 0
    for i in range(2):
        for j in range(2):
            key2d[i][j] = ord(key[iter3])-65
            iter3 += 1

#print (key2d)

    # checking validity of the key
    # finding determinant
    determnt = key2d[0][0] * key2d[1][1] - key2d[0][1] * key2d[1][0]
    determnt = determnt % 26

    # finding multiplicative inverse and implementing steps to encrypt text
    mult_invr = -1 
    for i in range(26):
        temp_invr = determnt * i
        if temp_invr % 26 == 1:
            mult_invr = i
            break
        else:
            continue
            
    if mult_invr == -1:
        print("Key is invalid")
        sys.exit()
    
    encryptd_text = ""
    iter_count = int(len(mssg)/2)
    if len_chkr == 0:
        for i in range(iter_count):
            temp1 = mssg2d[0][i] * key2d[0][0] + mssg2d[1][i] * key2d[0][1]
            encryptd_text += chr((temp1 % 26) + 65)
            temp2 = mssg2d[0][i] * key2d[1][0] + mssg2d[1][i] * key2d[1][1]
            encryptd_text += chr((temp2 % 26) + 65)
    else:
            for i in range(itr_count-1):
                temp1 = mssg2d[0][i] * key2d[0][0] + mssg2d[1][i] * key2d[0][1]
                encryptd_text += chr((temp1 % 26) + 65)
                temp2 = mssg2d[0][i] * key2d[1][0] + mssg2d[1][i] * key2d[1][1]
                encryptd_text += chr((temp2 % 26) + 65)
    
 

    print("Encrypted text: {}".format(encryptd_text))
   # return encryp_text


def cipher_decryption():
    mssg = input("Enter the message: ").upper()
    mssg = mssg.replace(" ", "")
    # Handle condition if the message length is odd.
    len_chkr = 0
    if len(mssg) % 2 != 0:
        mssg += "0"
        len_chkr = 1

    # Convert msg to matrices
    row = 2
    col = int(len(mssg) / 2)
    mssg2d = np.zeros((row, col), dtype=int)

    itr1 = 0
    itr2 = 0
    for i in range(len(mssg)):
        if i % 2 == 0:
            mssg2d[0][itr1] = int(ord(mssg[i]) - 65)
            itr1 += 1
        else:
            mssg2d[1][itr2] = int(ord(mssg[i]) - 65)
            itr2 += 1
            
               
    key = input("Enter the 4 letter key string please: ").upper()
    key = key.replace(" ", "")
    # Convert key to 2x2
    key2d = np.zeros((2, 2), dtype=int)
    iter3 = 0
    for i in range(2):
        for j in range(2):
            key2d[i][j] = ord(key[iter3])-65
            iter3 += 1

    # finding determinant
    determnt = key2d[0][0] * key2d[1][1] - key2d[0][1] * key2d[1][0]
    determnt = determnt % 26

    # finding multiplicative inverse
    mult_invr = -1 
    for i in range(26):
            temp_invr = determnt * i
            if temp_invr % 26 == 1:
                mult_invr = i
                break
            else:
                continue

    # adjugate matrix
    # swapping
    key2d[0][0], key2d[1][1] = key2d[1][1], key2d[0][0]

    #changing signs
    key2d[1][0] *= -1
    key2d[0][1] *= -1

    key2d[1][0] = key2d[1][0] % 26
    key2d[0][1] = key2d[0][1] % 26 


    # multiplying multiplicative inverse with adjugate matrix
    #stop the range at 2 since it is hill 2x2 technique
    for i in range(2):
        for j in range(2):
            key2d[i][j] *= mult_invr

    # Calculate modulo
    #to calculate the modulo, I set value of both the rows and the columns 
    # of the 2 dimensional array to the value of the  the rows and the columns 
    # of the 2 dimensional array divided by 26 and return the remainder.
    for i in range(2):
        for j in range(2):
            key2d[i][j] =  key2d[i][j] % 26

    # Convert cipher to plaintext
    #Now I will convert the cipher text to plaintext
    decryptd_text = ""
    iter_count = int(len(mssg)/2)
    if len_chkr == 0:
        for i in range(iter_count):
            temp1 = mssg2d[0][i] * key2d[0][0] + mssg2d[1][i] * key2d[0][1]
            decryptd_text += chr((temp1 % 26) + 65)
            temp2 = mssg2d[0][i] * key2d[1][0] + mssg2d[1][i] * key2d[1][1]
            decryptd_text += chr((temp2 % 26) + 65)
    else:
            for i in range(itr_count - 1):
                temp1 = mssg2d[0][i] * key2d[0][0] + mssg2d[1][i] * key2d[0][1]
                decryptd_text += chr((temp1 % 26) + 65)
                temp2 = mssg2d[0][i] * key2d[1][0] + mssg2d[1][i] * key2d[1][1]
                decryptd_text += chr((temp2 % 26) + 65)

    print("Decrypted text: {}".format(decryptd_text))
    
def main():
    userch = int(input("1. Encryption\n2. Decryption\nChoose(1,2): "))
    if userch == 1:
        print("***ENCRYPTION***")
        cipher_encryption()
    elif userch == 2:
        print("***DECRYPTION***")
        cipher_decryption()
    else:
        print("Invalid Choice Detected.")

if __name__ == "__main__":
    main()
```
